Creating a decision Tree classifier for the Iris dataset and visualising it graphically

### Data Preprocessing

```python
# Importing the essential libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset
data = pd.read_csv("Iris.csv")
data = data.iloc[:,1:]
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
# finding the count of differnt output values and visualizing it

print(data["Species"].value_counts())

plt.figure(figsize=(10,6))
plt.hist(data["Species"])
plt.xlabel("Species")
plt.ylabel("No.")
plt.show()
```

```python
# analyzing the correlationg between different features

plt.figure(figsize=(18,12))
sns.heatmap(data.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

```python
# plotting the graph between differnt features

sns.pairplot(data)
```

```python
# seperating as input and output values

X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
# label encoding the output

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
# performing train test split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

### Training

```python
# importing the decision tree classifier and creating the model for training 

from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='gini') 
```

```python
# training the model with the help of training set 

classifier.fit(x_train, y_train)
```

```python
y_train
```

### Prediction

```python
# prediction for the test data

y_pred = classifier.predict(x_test)
y_pred = le.inverse_transform(y_pred)
y_test = le.inverse_transform(y_test)
```

```python
y_pred
```

```python
# calculating the accuracy of the model
from sklearn.metrics import accuracy_score, confusion_matrix

# accuracy score

ac = accuracy_score(y_test, y_pred)
print("Accuracy is : ", ac)

# confusion matrix

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix :\n ", cm)
```

### Visualising the model graphically

```python
from sklearn import tree

plt.figure(figsize=(25,20))
dot_data = tree.plot_tree(classifier, 
                                feature_names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'],  
                                class_names=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
                                filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```
